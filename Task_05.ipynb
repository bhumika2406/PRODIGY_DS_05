```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Road Accident Data.csv")
df.head()
```

```python
print(df.info())
print(df.columns)
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10,5))
sns.countplot(
    data=df,
    x='Weather_Conditions',
    order=df['Weather_Conditions'].value_counts().index
)
plt.xticks(rotation=90)
plt.title("Accidents by Weather Conditions")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x='Road_Surface_Conditions',
    order=df['Road_Surface_Conditions'].value_counts().index
)

plt.xticks(rotation=45)
plt.title("Road Surface Conditions")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Light_Conditions',
    order=df['Light_Conditions'].value_counts().index
)

plt.xticks(rotation=45)
plt.title("Light Conditions During Accidents")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Day_of_Week',
    order=df['Day_of_Week'].value_counts().index
)

plt.title("Accidents by Day")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Speed_limit',
    order=df['Speed_limit'].value_counts().index
)

plt.title("Accidents by Speed Limit")
plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(
    data=df,
    x='Urban_or_Rural_Area'
)

plt.title("Urban vs Rural Accidents")
plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(
    data=df,
    x='Accident_Severity'
)

plt.title("Accident Severity")
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df['Longitude'],
    df['Latitude'],
    alpha=0.5
)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Accident Hotspots")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Number_of_Casualties'], bins=20)

plt.title("Distribution of Casualties")
plt.show()
```

```python
print("Most accidents happen during:")
print(df['Weather_Conditions'].value_counts().head())

print("\nMost common road condition:")
print(df['Road_Surface_Conditions'].value_counts().head())

print("\nMost accident-prone speed limit:")
print(df['Speed_limit'].value_counts().head())
```
